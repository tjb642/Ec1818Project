Next we create classes for the companies and applicants in the model.

```python
#This class includes all the data and methods need to create companies in the simulation.
class Company:
    def __init__(self, funds, recruit_needs):
        
        #Funds will be drawn from a normal distribution.
        self.start_sigma = funds*0.15
        self.funds = float(funds) + random.gauss(0, self.start_sigma)
        self.start_funds = self.funds
        self.recruit_needs = recruit_needs
        
        #Company type is 1 if they hunt only for top talent
        self.company_type = 1
        
        #Last applicant seen.
        self.last_seen = None
        
        #Records how many applicants have been recruited.
        self.recruit_counter = 0
        
        #Equals 1 if the company has reached an exit-simulation condition.
        self.done = 0
        
        #Total number of applicants seen.
        self.applicants_seen = 0
        
        #simulates aggressive vs. timid companies
        self.aggression = random.uniform(0.1, 0.25)
        
    def update_last_seen(self):
        self.last_seen += 1
        
    def print_info(self):
        print "type: " + str(self.company_type)
        print "funds: " + str(self.funds)
        print "recruiting needs: " + str(self.recruit_needs)
        print "recruit total: " + str(self.recruit_counter)
```

```python
#Contains all the methods and data needed to create applicants.
class Applicant:
    def __init__(self, talent, funds, recruit_needs):
        
        #Equal to 1 if considered "top talent".
        self.talent = talent
        
        #Dollar amount of last offer.
        self.offer = 0
        
        #Tracks number of offers received.
        self.offer_count = 0
        
        #Equal to 1 if offer has been accepted.
        self.offer_accept = 0
        
        #Simulates inherent value of an applicant. Value chosen from a normal function.
        if talent == 1:
            self.value = random.gauss(2000, (0.15 * 2000))
        else:
            
            #Non top talent only worth 75% of top talent.
            self.value = random.gauss(2000 * 0.75, (0.15 * (2000 * 0.75)))
        
        #Keeps track of companies making offers in case of bidding war.
        self.companies_interested = set()
        
    def update_companies_interested(self, company, inc):
        if inc == 1:
            self.companies_interested.add(company)
        else:
            self.companies_interested.remove(company)
    
    def print_info(self):
        print "type: " + str(self.talent)
        print "value: " + str(self.value)
        print "offer: " + str(self.offer)
        print "offer count: " + str(self.offer_count)
        print "offer accept: " + str(self.offer_accept)
```

And now two small functions to create a list of companies and a list of applicants.

```python
"""
create_companies 

Creates a list of companies given the parameters.

Inputs:
n - Integer. Number of companies desired to be created.
funds - Integer. Starting money for each company for recruiting.
recruit_needs - Integer. Number of recruits needed for each company.
percent_top_seeker - Float. Percent of companies that search only for top talent.

Outputs:
companies - List of companies.
"""

def create_companies(n, funds, recruit_needs, percent_top_seeker):
    companies = []
    for each in xrange(n):
        
        #Create the company
        x = Company(funds, recruit_needs)
        
        #The first n*percent_top_seeker are created as top talent seekers. After that regular companies.
        if each >= (n*percent_top_seeker):
            x.company_type = 0
            
            #If not a top talent seeker, the company will need to higher more people to make up for the lack of talent.
            x.recruit_needs = round((4/3) * recruit_needs)
        companies.append(x)
    return companies
```

```python
"""
create_applicants

Creates a list of applicants given the parameters.

Inputs:
n - Integer. Number of applicants desired.
percent_top_talent - Float. Percentage of applicants who are top talent.
funds - Integer. Starting funds for companies in the simulation.
recruit_needs - Integer. Recruiting needs for companies in the simulation.
"""

def create_applicants(n, percent_top_talent, funds, recruit_needs):
    applicants = []
    for each in xrange(n):
        
        #By drawing a random number from the uniform distribution and multiplying it by the percent_top_talent we
        #can randomly disperse top talent applicants throughout the list.
        rand = random.randint(1, 100)
        if rand <= (percent_top_talent * 100):
            x = Applicant(1, funds, recruit_needs)
            applicants.append(x)
            continue
        
        #Otherwise just a regular applicant.
        x = Applicant(0, funds, recruit_needs)
        applicants.append(x)
            
    return applicants
```

## Making the Simulation Work

First we need to create a function that governs how offers are made and received.

```python
def make_offer(company, applicant):
    global loop_break_sentinal
    if applicant.offer_accept == 1:
        company.update_last_seen()
        company.applicants_seen += 1
        return
    
    if applicant.talent == 1 and company.company_type == 1:
        #A company shows interest
        applicant.update_companies_interested(company, 1)
        
        #The company makes the offer. In this case we first check to make sure only 1 company is interested
        if len(applicant.companies_interested) == 1:
            
            #No offers yet so go by the inherent value
            if applicant.offer_count == 0:
                
                #Check to see if the company can afford the applicant
                if applicant.value + (company.aggression * applicant.value) > company.funds:
                    company.update_last_seen()
                    company.applicants_seen += 1
                    return
                
                #Make the offer
                applicant.offer = applicant.value + (company.aggression * applicant.value)
                applicant.offer_count += 1
                
                #Cost of interviews and actually putting together and making the offer
                company.funds -= (applicant.value * .01)
                
            #Offers exist so go by last offer
            else:
                #Check to see if the company can afford the applicant
                if applicant.offer + (company.aggression * applicant.offer) > company.funds:
                    company.update_last_seen()
                    company.applicants_seen += 1
                    return
                
                #Make the offer
                applicant.offer = applicant.offer + (company.aggression * applicant.offer)
                applicant.offer_count += 1
                
                #Cost of interviews and actually putting together and making the offer
                company.funds -= (applicant.value * .01)
            
            #If no other companies are interested, applicant gives up after 3 offers
            if applicant.offer_count == 3:
                applicant.offer_accept = 1
                company.funds -= applicant.offer
                company.recruit_counter += 1
                if company.recruit_counter == company.recruit_needs:
                    loop_break_sentinal += 1
                    company.done = 1
                company.update_last_seen()
                company.applicants_seen += 1
                return
            return
        
        if len(applicant.companies_interested) > 2:
            company.update_last_seen()
            company.applicants_seen += 1
            applicant.update_companies_interested(company, 0)
            return
        
        max_bid = applicant.value + (applicant.value * 2.5 * company.aggression)
        if applicant.offer > max_bid:
            company.update_last_seen()
            company.applicants_seen += 1
            applicant.update_companies_interested(company, 0)
            applicant.offer_accept = 1
            other_company = list(applicant.companies_interested)[0]
            other_company.funds -= applicant.offer
            other_company.recruit_counter += 1
            if other_company.recruit_counter == other_company.recruit_needs:
                loop_break_sentinal += 1
                other_company.done = 1
            return
        
        #Check to see if company can afford the applicant
        if applicant.offer + (company.aggression * applicant.offer) > company.funds:
            company.update_last_seen()
            company.applicants_seen += 1
            applicant.update_companies_interested(company, 0)
            return
        
        applicant.offer = applicant.offer + (company.aggression * applicant.offer)
        applicant.offer_count += 1
                
        #Cost of interviews and actually putting together and making the offer
        company.funds -= (applicant.value * 0.01)
        return
    
    elif applicant.talent == 0 and company.company_type == 0:
        #Check to see if company can afford the applicant
        if applicant.value > company.funds:
            company.update_last_seen()
            company.applicants_seen += 1
            return
        
        applicant.offer = applicant.value
                
        #Cost of interviews and actually putting together and making the offer
        company.funds -= (applicant.value * 0.01)
        
        applicant.offer_accept = 1
        company.funds -= applicant.offer
        company.recruit_counter += 1
        if company.recruit_counter == company.recruit_needs:
            loop_break_sentinal += 1
            company.done = 1
        company.update_last_seen()
        company.applicants_seen += 1
        return
    
    company.update_last_seen()
    company.applicants_seen += 1
    return
```

```python
def market_sim(companies, applicants):
    global loop_break_sentinal
    loop_break_sentinal = 0
    while True:
        for company in companies:
            if company.done == 1:
                continue
            
            if company.last_seen is None:
                company.last_seen = random.randint(0, (len(applicants) - 1))
                first_seen = company.last_seen
                make_offer(company, applicants[first_seen])
                if company.last_seen == len(applicants):
                    company.last_seen = 0
                continue
            
            last_seen = company.last_seen
            make_offer(company, applicants[last_seen])
            
            if company.applicants_seen == len(applicants):
                loop_break_sentinal += 1
                company.done = 1
                continue
            
            if company.last_seen == len(applicants):
                company.last_seen = 0
                
        if loop_break_sentinal == len(companies):
            return False
```

```python
def compute_score(company, funds):
    #score equals total funds leftover + recruits recruited - recruit spaces unfilled
    #future implementation: include "buyer's remorse" for overvaluation effect
    if company.company_type == 1:
        score = company.funds/(funds * 0.1) + (4/3) * company.recruit_counter - (4/3) * (company.recruit_needs
                                                                                         - company.recruit_counter)
    else:
        score = company.funds/(funds * 0.1) + company.recruit_counter - (company.recruit_needs - company.recruit_counter)
    
    return score
```

```python
def complete_sim():
    company_n = [5, 10, 50, 100]
    funds = [1000, 5000, 10000, 100000]
    needs = [10, 50, 100, 500]
    company_perc = [.25, .5, .75, .9]
    applicant_n = [10000, 25000, 50000, 100000]
    applicant_perc = [.01, .05, .10, .15]
    results = []
    for n in company_n:
        df = pd.DataFrame(index = xrange(n))
        for fund in funds:
            print "Okay so far"
            for need in needs:
                for perc in company_perc:
                    for m in applicant_n:
                        for aperc in applicant_perc:
                            companies = create_companies(n, fund, need, perc)
                            applicants = create_applicants(m, aperc, fund, need)
                            market_sim(companies, applicants)
                            heading = str(n) + "," + str(fund) + "," + str(need) + "," + str(perc) + "," + str(m) + "," + str(aperc)
                            temp = []
                            types = []
                            for company in companies:
                                temp.append(compute_score(company, fund))
                                types.append(company.company_type)
                            temp = np.array(temp)
                            types = np.array(types)
                            df[heading] = temp
                            df[heading + "Types"] = types
        results.append((n, df))
    return results
```
